```python
a = int(input())
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
uoc = []
for n in range(2, a):
    if a % n == 0 and is_prime(n):
        uoc.append(n)
    if n == a:
        break
print(max(uoc))
```
