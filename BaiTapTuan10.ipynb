```python
def quick_sort(n):
    if len(n) <= 1:
        return n
    left = []
    right = []
    middle = []
    pivot = n[len(n) // 2]
    for x in n:
        if x < pivot:
            left.append(x)
        elif x > pivot:
            right.append(x)
        else:
            middle.append(x)
    return sort(left, middle, right)
def sort(left, middle, right):
    return quick_sort(left) + middle + quick_sort(right)
m = list(map(int, input().split()))
n = list(map(int, input().split()))
print(quick_sort(m+n))
```

```python
def sapxep(n):
    if len(n) <= 1:
        return n
    pivot = min(n)
    if n[0] != pivot:
        n.insert(n.index(pivot), n[0])
        n.remove(n[0])
        n.remove(pivot)
        n.insert(0, pivot)
        return n
    elif n[0] == pivot:
        n.insert(len(n)-2, n[-1])
        n.pop(n[-1])
        return n
n = input()
a = []
c = 0
for x in n:
    a.append(int(x))
for y in sapxep(a):
    c = c*10 +y
print(c)
```

```python
n = input()
m = input()
c = 0
if len(m) > len(n):
    print('NO')
else:
    for x in n:
        if x == m[0] or ord(x) - 32 == ord(m[0]):
            b = n[n.index(x):]
    if len(b) > 0:
        for i in range(0, len(m)):
            if b[i] == m[i] or ord(b[i])-32 == ord(m[i]):
                c += 1
    elif len(b) == 0:
        print('NO')
if c == len(m):
    print('YES')
else:
    print('NO')
```
