```python
x = float(input("diem khoi tao"))
import math
y = 3 * math.cos(x) - x -1
dy = -3 * math.sin(x) - 1
x = x - (y / dy)
while abs(y - 10**-10) > 10**-10:
    y = 3 * math.cos(x) - x -1
    dy = -3 * math.sin(x) - 1
    x = x - (y / dy)
    if abs(y - 10**-10) <= 10**-100:
        break
print(x)
```
