```python
a = list(map(int, input().split()))
x = int(input())
for i in a:
    if i < x:
        a.remove(i)
print(a)
```

```python
a = list(map(int, input().split()))
b = []
for x in a:
    if x != 0:
        b.append(x)
for _ in range(len(a)-len(b)):
    b.insert(0, 0)
print(b)
```
