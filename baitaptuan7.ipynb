```python
def quick_sort(n):
    if len(n) <= 1:
        return n
    left = []
    right = []
    middle = []
    pivot = n[len(n)//2]
    for x in n:
        if x < pivot:
            left.append(x)
        elif x > pivot:
            right.append(x)
        else:
            middle.append(pivot)
    return sort(left, middle, right)
def sort(left, middle, right):
    return quick_sort(left) + middle + quick_sort(right)
n = list(map(int, input().split()))
print(quick_sort(n))
```
