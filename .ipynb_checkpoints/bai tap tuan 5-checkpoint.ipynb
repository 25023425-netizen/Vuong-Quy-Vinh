```python
def prime(n):
    for i in range (2,n):
        if n % i == 0:
            break
        return False
    return True
```

```python
def perfect(n):
    a = []
    tong_uoc = 0
    for i in range (2,n):
        if n % i == 0:
            a.append(i)
            for x in a
                tong_uoc += i
    if tong_uoc == n:
        return True
    return False
```
