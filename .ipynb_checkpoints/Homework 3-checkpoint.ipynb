```python
#bai 7
n = int(input())
S=0
print("So uoc cua",n,"la")
for i in range(1,n+1):
    a = n/i
    if a.is_integer():
        S = S+1
    if i == n:
        break
print(S)
```

```python
#bai 8
a = int(input())
b = int(input())
print("cac uoc chung la")
if a <= b:
    for i in range(1, b+1):
        e = a/i
        f = b/i
        if e.is_integer() and f.is_integer():
            print(i)
if a > b:
    for i in range(1, a+1):
        e = a/i
        f = b/i
        if e.is_integer() and f.is_integer():
            print(i)
```
