```python
n = list(map(int, input().split()))
A = {"positives":[], "negatives":[], "zero":[]}
for x in n:
    if x < 0:
        A["negatives"].append(x)
    if x > 0:
        A["positives"].append(x)
    if x == 0:
        A["zero"].append(x)
print(A)
```

```python
n = list(map(int, input().split()))
b = list(set(n))
c = {}
for i in range(len(b)):
    d = 0
    for _ in range(len(n)):
        if b[i] == n[_]:
            d += 1
    key = b[i]
    c[key] = d
print(c)
```

```python
a = input("nhap vao cac cap key:value")
b = input("nhap vao cac cap key:value")
c = {}
A = a.split()
B = b.split()
for x in A:
    key, value=x.split(':', 1)
for y in B:
    key1, value1=y.split(':', 1)
if key in c:
    c[key].append(value)
    if key1 in c:
        c[key].append(value1)
else:
    c[key1]=[value1]
print(c)
```

```python
def dict(n):
    d = {}
    n = n.split()
    for x in n:
        key, value=x.split(':', 1)
        if key in d:
            d[key].append(value)
        else:
            d[key] = [value]
    return d
a = input()
b = input()
print(sorted(dict(a+' '+b)))
```

```python
a = list(map(int, input().split()))
a = sorted(a)
b = []
k = int(input())
for i in range (0, len(a)):
    for _ in range (i+1, len(a)):
        if a[i] + a[_] == k:
            key = sorted([a[i], a[_]])
            b.append(tuple(key))
print(list(set(b)))
```

```python
def sochan(n):
    if n%2 == 0:
        return True
    return False
n = list(map(int, input().split()))
a = []
b = []
for x in n:
    if sochan(x):
        a.append(x)
    else:
        b.append(x)
print(tuple(a))
print(tuple(b))
```
