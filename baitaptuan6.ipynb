```python
n = list(map(int, input().split()))
x = 0
y = 0
z = 0
for i in n:
    if i == 0:
        z = z + 1
    elif i>0:
        x = x + 1
    elif i<0:
        y = y + 1
print({'positives':x , 'negatives':y, 'zeros':z})
```
